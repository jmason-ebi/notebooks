```python
import datetime, calendar

NOW = datetime.datetime.now()
year = NOW.year
month = NOW.month
num_days = calendar.monthrange(year, month)[1]
days = [datetime.date(year, month, day) for day in range(1, num_days+1)]
days = sorted(days, reverse=True)

print ("""<h1>%s</h1>
<p class="auto-cursor-target">
  <br/>
</p>
<table class="wrapped">
  <colgroup>
    <col/> <col/> <col/> </colgroup>
  <tbody>
    <tr>
      <th>Date</th>
      <th>Day</th>
      <th>Activity</th>
    </tr>""" % (NOW.strftime("%B %Y")))
for day in days:
    print ("""     <tr>
      <td>%s</td>
      <td>%s</td>
      <td>
        <br/>
      </td>
    </tr> """ % (day.strftime("%Y%m%d"), day.strftime("%A")))

print ("""  </tbody>
</table>""")
```
